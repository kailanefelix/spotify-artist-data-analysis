```python
import spotipy
import csv
import pandas as pd
```

```python
from spotipy.oauth2 import SpotifyClientCredentials

sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id="2f165a1bdfe941a691045509e35ae587",
                                                           client_secret="4e1d5e13ff9142078bd4cd02bd03559a"))
```

```python
# searching by name (return the summary informations about the artist)

def get_artist(name):
    search = sp.search(name, type='artist')
    items = search['artists']['items']
    if items:
        return items[0]
```

```python
# searching albums by artist id

def get_albums_data(artist_id):
    albums = {}
    artist_albums = sp.artist_albums(artist_id, limit=25)

    for item in artist_albums['items']:
        albums[item['name']] = item
        
    return albums.values()


albums_data = get_albums_data('66CXWjxzNUsdJxJ2JdwvnR')
for album in albums_data:
    print('name:', album['name'], 'id:', album['id'])
```

```python
# searching tracks by album
def get_tracks(album_id):
    spotify_album = {}
  
    tracks = sp.album_tracks(album_id)
  
    for track in tracks['items']:
        id_track = track['id']
        track = sp.track(id_track)
        track_name = track['name']
        
        spotify_album[track_name] = {
            'id': track['id'], 
            'track_number': track['track_number'],
            'popularity': track['popularity'],  
       }
    
    return spotify_album

pd.DataFrame(get_tracks('3WSg3hc6uZ7AYZBvtsg2GU')).T
```
